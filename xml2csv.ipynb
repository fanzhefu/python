# 1 Convert XML to CSV Using Python
https://www.delftstack.com/howto/python/xml-to-csv-python/

```python
student_data_as_string = '''
<students>
    <student>
        <name>Rick Grimes</name>
        <rollnumber>1</rollnumber>
        <age>15</age>
    </student>
    <student>
        <name>Lori Grimes</name>
        <rollnumber>2</rollnumber>
        <age>16</age>
    </student>
    <student>
        <name>Carl Grimes</name>
        <rollnumber>3</rollnumber>
        <age>14</age>
    </student>
    <student>
        <name>Judith Grimes</name>
        <rollnumber>4</rollnumber>
        <age>13</age>
    </student>
</students>
'''
```

```python
import xml.etree.ElementTree as ET

root = ET.fromstring(student_data_as_string)
# tree = ElementTree.parse("input.xml")
# root = tree.getroot()
# root = tree.getroot()

for student in root:
    name = student.find("name").text
    roll_number = student.find("rollnumber").text
    age = student.find("age").text
    print(f"{name},{roll_number},{age}")
```

# 2 Convert XML to CSV in Python
https://www.geeksforgeeks.org/convert-xml-to-csv-in-python/

```python
# Importing the required libraries
import xml.etree.ElementTree as ET
import pandas as pd

# cols = ["name", "phone", "email", "date", "country"]
cols = ["name", "rollnumber", "age"]
rows = []

# Parsing the XML file
# xmlparse = Xet.parse('sample.xml')
# root = xmlparse.getroot()
root = ET.fromstring(student_data_as_string)

for i in root:
	name = i.find("name").text
	rollnumber = i.find("rollnumber").text
	age = i.find("age").text
# 	date = i.find("date").text
# 	country = i.find("country").text

	rows.append({"name": name,
				"rollnumber": rollnumber,
				"age": age
                })

df = pd.DataFrame(rows, columns=cols)

# Writing dataframe to csv
# df.to_csv('output.csv')
```
