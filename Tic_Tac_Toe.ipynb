```python
# 
import os
import random

class Tic_Tac_Toe():
    '''Human X-man vs Odinateur Ordi'''
    def __init__(self):
        self.board = [' ' for x in range(9)]
        self.spots = {0,1,2,3,4,5,6,7,8}
        self.play = True # set to False if win or tie.
        
    def print_board(self):
        '''Display the board.'''
        print(' ' + self.board[0] + ' | ' + self.board[1] + ' | ' + self.board[2])
        print(' ' + self.board[3] + ' | ' + self.board[4] + ' | ' + self.board[5])
        print(' ' + self.board[6] + ' | ' + self.board[7] + ' | ' + self.board[8])
        
    def insert_letter(self, letter, position): # X or O
        '''Insert "X" or "O" on the board.'''
        self.board[position] = letter

    def is_space_free(self, position): # return T/F
        '''Check if the position is free.'''
        return self.board[position] == ' '

    def is_board_full(self): # return T/F
        '''Is board full.'''
        return False if self.board.count(' ') > 0 else True

    def is_winner(self, letter): # return T/F 
        '''If player "X" or "O" win the game'''
        return (self.board[0] == letter and self.board[1] == letter and self.board[2] == letter) or \
               (self.board[3] == letter and self.board[4] == letter and self.board[5] == letter) or \
               (self.board[6] == letter and self.board[7] == letter and self.board[8] == letter) or \
               (self.board[0] == letter and self.board[3] == letter and self.board[6] == letter) or \
               (self.board[1] == letter and self.board[4] == letter and self.board[7] == letter) or \
               (self.board[2] == letter and self.board[5] == letter and self.board[8] == letter) or \
               (self.board[0] == letter and self.board[4] == letter and self.board[8] == letter) or \
               (self.board[2] == letter and self.board[4] == letter and self.board[6] == letter)
   
#     def select_random(lst):
#         n = random.randrange(0,len(lst))
#         return lst[n]
    def xman_move(self):
        move = 9 # 9 is returned if board is full
        if not self.is_board_full():
            run = True
            while run:
                move = input("Select a number of you choice and that will be your block: \'X\' (0-8): ")
                try:
                    move = int(move)
                    if 0 <= move <= 8:
                        if self.is_space_free(move):
                            run = False
                            #insert_letter('X', move)
                        else:
                            print("Choose another block, it's been taken!")
                    else:
                        print('Please type a number within the range(0-8)!')
                except:
                    print('Please type a number(0-8)!')
        #return move
        self.insert_letter('X', move)
        self.print_board()
        self.spots.remove(move)
        if self.is_winner('X') or len(self.spots)==0:
            self.play = False 
            
            
    def ordi_move(self): # return move position, 9 for board full
        '''"O" ordi check for the winning position first if not, then move randomly'''
        
        #find index of the ' ' in the borad list
        possible_moves = [i for i, letter in enumerate(self.board) if letter == ' ']

        move = 9 # 9 is not a valide move, 9 returned if board is full
        
        # find the wining move 
        for letter in ['O', 'X']:
            for i in possible_moves:
                self.board[i] = letter
                if self.is_winner(letter):
                    move = i
                    self.board[i] = ' '
                    break
#                     return move
                self.board[i] = ' '

        #if four corners 0,2,6,8 are open, do the corner move.
        corners_open = []
        for i in possible_moves:
            if i in [0,2,6,8]:
                corners_open.append(i)
        if len(corners_open) > 0:
            move = random.choice(corners_open)
#             return move

        #then do the centre move 
        if 4 in possible_moves:
            move = 4
#             return move

        #four edges 1,3,5,7 move
        edges_open = []
        for i in possible_moves:
            if i in [1,3,5,7]:
                edges_open.append(i)
        if len(edges_open) > 0:
            move = random.choice(edges_open)
#             return move
        
        self.insert_letter('O', move)
        self.print_board()
        self.spots.remove(move)
        if self.is_winner('O') or len(self.spots)==0:
            self.play = False 
            

def game_on():
    print('Welcome to Tic Tac Toe!', "\nEnter 'yes' if you (X-man) want to start the game or 'no' for me (Odinateur) to start it." )
    start = input("Yes or No:")
    ttt = Tic_Tac_Toe()
    ttt.print_board()
    print(start)
    if start.lower() == 'yes'  or start.lower == 'y':  # X-man goes first
        while ttt.play:
            print('x-move')
            ttt.xman_move()
            print('o-move')
            ttt.ordi_move()
            
    elif start.lower() == 'no' or start.lower == 'n': # Odinateur goes first
        pass
    
    else:
        print("\nWrong input !!!")

# Game On ======================================================================
print('''
Congratulations on reaching this far. You deserve a treat now! You deserve to play a game, but it is not that simple!! In order to have a fun time playing the game, you need to create one!!! Let's creat Tic-Tac-Toe.

0|1|2
3|4|5
6|7|8

We will see if you are able to win the game against me.

Do not worry... I will help you. You will make you move by entering a number between 0 - 8. The number will represent the position on the playing board. The board may look similiar to one shown above. I challenge you to beat me!!!!
''')  

# loop for play or quit
while True:
#     board = [' ' for i in range(9)]
    print('-----------------------------------')
    game_on()
    answer = input('Do you want to play again? (Y/N)')
    #clear()
    if answer.lower() == 'n' or answer.lower == 'no':
        break
        
```

```python
#Tic Tac Toe game in python by techwithtim

board = [' ' for x in range(10)]

def insertLetter(letter, pos):
    board[pos] = letter

def spaceIsFree(pos):
    return board[pos] == ' '

def printBoard(board):
    print('   |   |')
    print(' ' + board[1] + ' | ' + board[2] + ' | ' + board[3])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[4] + ' | ' + board[5] + ' | ' + board[6])
    print('   |   |')
    print('-----------')
    print('   |   |')
    print(' ' + board[7] + ' | ' + board[8] + ' | ' + board[9])
    print('   |   |')
    
def isWinner(bo, le):
    return (bo[7] == le and bo[8] == le and bo[9] == le) or (bo[4] == le and bo[5] == le and bo[6] == le) or(bo[1] == le and bo[2] == le and bo[3] == le) or(bo[1] == le and bo[4] == le and bo[7] == le) or(bo[2] == le and bo[5] == le and bo[8] == le) or(bo[3] == le and bo[6] == le and bo[9] == le) or(bo[1] == le and bo[5] == le and bo[9] == le) or(bo[3] == le and bo[5] == le and bo[7] == le)

def playerMove():
    run = True
    while run:
        move = input('Please select a position to place an \'X\' (1-9): ')
        try:
            move = int(move)
            if move > 0 and move < 10:
                if spaceIsFree(move):
                    run = False
                    insertLetter('X', move)
                else:
                    print('Sorry, this space is occupied!')
            else:
                print('Please type a number within the range!')
        except:
            print('Please type a number!')
            

def compMove():
    possibleMoves = [x for x, letter in enumerate(board) if letter == ' ' and x != 0]
    move = 0

    for let in ['O', 'X']:
        for i in possibleMoves:
            boardCopy = board[:]
            boardCopy[i] = let
            if isWinner(boardCopy, let):
                move = i
                return move

    cornersOpen = []
    for i in possibleMoves:
        if i in [1,3,7,9]:
            cornersOpen.append(i)
            
    if len(cornersOpen) > 0:
        move = selectRandom(cornersOpen)
        return move

    if 5 in possibleMoves:
        move = 5
        return move

    edgesOpen = []
    for i in possibleMoves:
        if i in [2,4,6,8]:
            edgesOpen.append(i)
            
    if len(edgesOpen) > 0:
        move = selectRandom(edgesOpen)
        
    return move

def selectRandom(li):
    import random
    ln = len(li)
    r = random.randrange(0,ln)
    return li[r]
    

def isBoardFull(board):
    if board.count(' ') > 1:
        return False
    else:
        return True

def main():
    print('Welcome to Tic Tac Toe!')
    printBoard(board)

    while not(isBoardFull(board)):
        if not(isWinner(board, 'O')):
            playerMove()
            printBoard(board)
        else:
            print('Sorry, O\'s won this time!')
            break

        if not(isWinner(board, 'X')):
            move = compMove()
            if move == 0:
                print('Tie Game!')
            else:
                insertLetter('O', move)
                print('Computer placed an \'O\' in position', move , ':')
                printBoard(board)
        else:
            print('X\'s won this time! Good Job!')
            break

    if isBoardFull(board):
        print('Tie Game!')

while True:
    answer = input('Do you want to play again? (Y/N)')
    if answer.lower() == 'y' or answer.lower == 'yes':
        board = [' ' for x in range(10)]
        print('-----------------------------------')
        main()
    else:
        break
```
